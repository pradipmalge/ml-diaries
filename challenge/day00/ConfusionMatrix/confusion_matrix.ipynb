# Confusion Matrix in Machine Learning

A **confusion matrix** is a table used to evaluate the performance of a classification algorithm. It shows the counts of actual vs predicted classifications, helping us understand where the model is getting things right or wrong.

## Parameters Explained

- **True Positive (TP):** Model correctly predicts the positive class.
- **True Negative (TN):** Model correctly predicts the negative class.
- **False Positive (FP):** Model incorrectly predicts positive when it is actually negative.
- **False Negative (FN):** Model incorrectly predicts negative when it is actually positive.

## Short Notes on Classification Metrics

- **Accuracy:**  
    The ratio of correctly predicted observations to the total observations.  
    $$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

- **Precision:**  
    The ratio of correctly predicted positive observations to the total predicted positive observations.  
    $$ \text{Precision} = \frac{TP}{TP + FP} $$

- **Recall (Sensitivity):**  
    The ratio of correctly predicted positive observations to all actual positives.  
    $$ \text{Recall} = \frac{TP}{TP + FN} $$

- **F1 Score:**  
    The harmonic mean of Precision and Recall.  
    $$ \text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$

- **Error Type I (False Positive):**  
    Predicting positive when it is actually negative (FP).

- **Error Type II (False Negative):**  
    Predicting negative when it is actually positive (FN).

```python
## Import necessary libraries

import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
## Actual and predicted values (Email Spam Classification Example)
actual = np.array([
    'Spam', 'Spam', 'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Spam', 'Not Spam', 'Not Spam'
])
predicted = np.array([
    'Spam', 'Not Spam', 'Spam', 'Not Spam', 'Spam', 'Spam', 'Spam', 'Spam', 'Not Spam', 'Not Spam'
])
```

```python
## Generate confusion matrix
cm = confusion_matrix(actual,predicted)
print("Confusion Matrix:\n", cm)
```

```python
## Plotting the confusion matrix
sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=['Spam','Not Spam'],
            yticklabels=['Spam','Not Spam'])
plt.ylabel('<---Actual--->', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('<----Prediction---->', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Email Prediction', ha='center', fontsize=15)
plt.show()
```

```python
## Classification report
print(classification_report(actual, predicted))
```
